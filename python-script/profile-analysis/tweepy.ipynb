```python
import tweepy
import os 
```

```python
with open("token2.txt", "r") as file:
    lines = file.read().splitlines()
    bearer_token = lines[0]
    api_key = lines[1]
    api_key_secret = lines[2]
    access_token = lines[3]
    access_token_secret = lines[4]
```

```python
client = tweepy.Client(bearer_token=bearer_token, 
                       consumer_key=api_key,
                       consumer_secret=api_key_secret,
                       access_token=access_token,
                       access_token_secret=access_token_secret)
```

```python
username = "haydarkadioglu_"  # Ä°nceleyeceÄŸiniz hesabÄ±n kullanÄ±cÄ± adÄ±
tweet_count = 100

# KullanÄ±cÄ± ID'sini alma
user = client.get_user(username=username)
user_id = user.data.id

# Son tweetleri alma
response = client.get_users_tweets(id=user_id, max_results=tweet_count, tweet_fields=["created_at"])

```

```python
tweets = []
for tweet in response.data:
    tweets.append({
    "Tweet ID": tweet.id,
    "Created At": tweet.created_at,
    "Tweet Text": tweet.text,
    })
```

```python
for tweet in tweets[:10]:
    print(tweet)
```

```python
import pandas as pd

# Convert the list of tweets to a DataFrame
df_tweets = pd.DataFrame(tweets)

# Save the DataFrame to a CSV file
df_tweets.to_csv('tweets.csv', index=False)
```
